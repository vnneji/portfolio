```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
```

```python
cancer_data.keys()
```

```python
df = pd.DataFrame(data = cancer_data['data'],columns = cancer_data['feature_names'], )
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(df)
```

```python
df_scaled = scalar.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(df_scaled)
```

```python
pc12 = pca.transform(df_scaled)
```

```python
type(pc12)
```

```python
df_pca = pd.DataFrame(data = pc12, columns = ['PC1', 'PC2'])
```

```python
df_pca.head()
```

```python
df_pca.info()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10,8))
plt.scatter(x=df_pca['PC1'], y=df_pca['PC2'], c = cancer_data['target'], cmap = 'Set1')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
pca.components_
```

```python
df_pca_comp = pd.DataFrame(data = pca.components_, columns = cancer_data['feature_names'])
```

```python
df_pca_comp.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data = df_pca_comp, cmap = 'rainbow')
```

```python
X=df_pca
y = cancer_data['target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(X_train, y_train)
pred = logr.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
print(confusion_matrix(y_test, pred))
```
