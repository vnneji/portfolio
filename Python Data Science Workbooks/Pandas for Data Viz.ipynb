```python
import numpy as np, pandas as pd
%matplotlib inline
```

```python
##Create dataframes
dates = pd.date_range("01-01-2000",periods = 500)
```

```python
col_D = np.random.rand(500)
```

```python
data_1 = np.random.randn(500,3)
```

```python
df1 = pd.DataFrame(data = data_1, index = dates, columns = ['A', 'B', 'C'])
```

```python
df1.head()
```

```python
df1['D'] = col_D
```

```python
df1.head()
```

```python
data_2 = np.random.rand(20,3)
col = ['X', 'Y', 'Z']
df2 = pd.DataFrame(data = data_2, columns = col)
df2.head()
```

```python
df2.plot.bar()
```

```python
##horizontal bar
df2.plot.barh()
```

```python
#Stacked bar
df2.plot.bar(stacked = True)
```

```python
##Histogram
df1['A'].plot.hist(bins = 30)
```

```python
df1.plot.hist(bins = 30, alpha = .5)
```

```python
##Stacked hist
df1.plot.hist(bins = 30, stacked = True)
```

```python
###Change the hist plot orientation
df1.plot.hist(bins = 30, stacked = True, orientation='horizontal')
```

```python
##Lineplot
df1.plot.line(y='A')
```

```python
##Formatting
df1.plot.line(y='A', lw = 1, figsize = (12,3))
```

```python
###Scatterplot
###Needs numerical values for x and y
df1.plot.scatter(x = 'A', y='B')
```

```python
#Color formating
ax = df1.plot.scatter(x = 'A', y='B', color = 'DarkBlue', label = 'Group1')
ax = df1.plot.scatter(x = 'C', y='D', color = 'DarkGreen', ax = ax, label = 'Group2')
```

```python
##Color plots based on column values 
df1.plot.scatter(x='A', y = 'B', c = 'C')
```

```python
##Change Scatter bubble size with color map (static size 50)
df1.plot.scatter(x='A', y = 'B', c = 'C', cmap = 'coolwarm', s = 50)
```

```python
##Change Scatter bubble size with color map (dynamic size based on column)
df1.plot.scatter(x='A', y = 'B', cmap = 'coolwarm', s = df1['C']*100)
```

```python
##Boxplot visualises the distribution of values within each column
df2.plot.box()
```

```python
##Change orientation
df2.plot.box(vert = False)
```

```python
##Passing colour dictionary
color = dict(boxes = 'DarkGreen', whiskers = 'DarkOrange', medians = 'DarkBlue', caps = 'Grey')
df2.plot.box(color = color)
```

```python
##Pie plot
###Nan and 0 will be replaced with 0
df2.plot.pie(y='X', figsize=(4,4), legend = False)
```
