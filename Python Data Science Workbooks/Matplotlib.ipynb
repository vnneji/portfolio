```python
import matplotlib.pyplot as plt
import numpy as np

###Set plots inline
%matplotlib inline
###Define values
x=np.arange(11)
y=x**2
```

```python
####Simple plot using the pyplot
plt.plot(x,y)
plt.xlabel('X axis ')
plt.ylabel('Y axis ')
plt.title('Figure x and x**2')
```

```python
#Compound plot using subplot (nrow, ncol, nplot)
plt.subplot(1,3,1)
plt.plot(x,y,'r--')

plt.subplot(1,3,2)
plt.plot(y,x,'g--')

plt.subplot(1,3,3)
plt.plot(x,y,'g*-')
plt.plot(y,x,'r*-')
plt.tight_layout()
```

```python
##Instantiate a Plot object using plt.figure()
fig1 = plt.figure()
axes = fig1.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
##Instantiate a Plot object using plt.figure()
###Create embedded plots using dual axes
fig2 = plt.figure()
axes_main = fig2.add_axes([0.1, 0.1, 0.9, 0.9])
axes_insert = fig2.add_axes([0.3, 0.5, 0.3, 0.4])
axes_main.plot(x,y,'b')
axes_insert.plot(y,x,'r')

axes_main.set_xlabel('Main X')
axes_main.set_ylabel('Main Y')
axes_main.set_title('Main Title')

axes_insert.set_xlabel('Insert X')
axes_insert.set_ylabel('Insert Y')
axes_insert.set_title('Insert Title')
```

```python
###Compound plot using subplots
fig, axes = plt.subplots(nrows = 2, ncols = 2, sharex=True, sharey=True)
plt.tight_layout()
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')

```

```python
fig1, axes1 = plt.subplots(nrows = 2, ncols = 1)
fig2, axes2 = plt.subplots(nrows = 1, ncols = 2)
```

```python
###Subplots axes are arrays
type(axes1),type(axes2)
```

```python
##Iterate through axes using for loop and set their properties
for ax in axes1:
    ax.plot(x,y)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
fig1.tight_layout()
fig1
```

```python
##Set axes properties using index notation
axes2[0].plot(x,y,'r')
axes2[0].set_xlabel('X')
axes2[0].set_ylabel('Y')
axes2[0].set_title('title')


axes2[1].plot(y,x,'g')
axes2[1].set_xlabel('X')
axes2[1].set_ylabel('Y')
axes2[1].set_title('title')

fig2.tight_layout()
fig2
```

```python
###Set plot object legend
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label = 'x**2')
axes.plot(x,x**3, label = 'x**3')
axes.legend(loc=0)
```

```python
##Design line color and patern
fig,axes = plt.subplots()
axes.plot(x,x+1, 'r.-')
axes.plot(x,x+2, 'g--')
```

```python
##Design colour and transparency
fig,axes = plt.subplots()
axes.plot(x,x+1, color = 'blue', alpha = 0.6)
axes.plot(x,x+2, color = '#8B008B')
axes.plot(x,x+3, color = '#FF8C00')
```

```python
fig, axes = plt.subplots(figsize = (10,8))
axes.plot(x,y,color = 'purple', lw = 4, ls = '-', marker = 's', 
          markersize = 18, markerfacecolor = 'yellow', markeredgewidth = 4,
         markeredgecolor = 'red')
```

```python
###Customise axes (limits)
fig, axes = plt.subplots(nrows=1, ncols = 3, figsize = (10,3))
axes[0].plot(x,y)
axes[0].set_title('default')

axes[1].plot(x,y)
axes[1].axis('tight')
axes[1].set_title('tight')

axes[2].plot(x,y)
axes[2].set_xlim([1,4])
axes[2].set_ylim([0,50])
axes[2].set_title('custom')
```
