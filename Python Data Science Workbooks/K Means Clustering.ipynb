```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
#create artificial clusters
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=1000,
    n_features=2,
    centers=4,
    cluster_std=3.0,
    random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data[0])
cent = kmeans.cluster_centers_
kmeans.labels_
fig, axes = plt.subplots(1,2,sharey=True, figsize = (10,6))
axes[0].scatter(x=data[0][:,0], y=data[0][:,1], c=kmeans.labels_, cmap = 'rainbow')
axes[0].scatter(x=cent[:,0],y=cent[:,1], s = 100, c='black')
axes[1].scatter(x=data[0][:,0], y=data[0][:,1], c=data[1], cmap = 'rainbow')
```

```python
sum_sq = {}
for k in range(1,10):
    kmeans = KMeans(n_clusters=k).fit(data[0])
    sum_sq[k] = kmeans.inertia_
```

```python
sum_sq
```

```python
plt.plot(list(sum_sq.keys()), list(sum_sq.values()), marker = '*', markersize = 10, markerfacecolor='red')
```
